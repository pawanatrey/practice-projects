```python
from sklearn.preprocessing import StandardScaler

x_scaler=StandardScaler()
x_scaler=x_scaler.fit_transform(x)
x_scaler
```

```python
# split the data

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error

x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.25,random_state=40)
```

```python
# define function 

def met_score(met,x_train1,x_test1,y_train1,y_test1,train=True):
    if train:
        y_pred=met.predict(x_train1)
        print("***************Train Result************")
        print("R2_Score", r2_score(y_train1,y_pred))
        print("Adjusted_R2_Score", met.score(x_train1,y_train1))        
        print("Mean Absolute Error", mean_absolute_error(y_train1,y_pred))
        print("Mean Square Error", mean_squared_error(y_train1,y_pred))
        print("Root Mean Square Error", np.sqrt(mean_squared_error(y_train1,y_pred)))
        
        
    elif train==False:
        pred=met.predict(x_test1)
        print(pred)
        print("***************Test Result************")
        
        print("R2_Score", r2_score(y_test1,pred))
        print("Adjusted_R2_Score", met.score(x_test1,y_test1))  
        print("Mean Absolute Error", mean_absolute_error(y_test1,pred))
        print("Mean Square Error", mean_squared_error(y_test1,pred))
        print("Root Mean Square Error", np.sqrt(mean_squared_error(y_test1,pred)))        
```

```python
# Linear Regression Model

from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
# call the function

met_score(lin_reg,x_train,x_test,y_train,y_test,train=True)

met_score(lin_reg,x_train,x_test,y_train,y_test,train=False)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
```

```python
met_score(knn,x_train,x_test,y_train,y_test,train=True)

met_score(knn,x_train,x_test,y_train,y_test,train=False)
```

```python
from sklearn.tree import DecisionTreeRegressor
Dec_Tree=DecisionTreeRegressor()
Dec_Tree.fit(x_train,y_train)
```

```python
met_score(Dec_Tree,x_train,x_test,y_train,y_test,train=True)

met_score(Dec_Tree,x_train,x_test,y_train,y_test,train=False)
```

```python
from sklearn.ensemble import AdaBoostRegressor

ada=AdaBoostRegressor()
ada.fit(x_train,y_train)
```

```python
met_score(ada,x_train,x_test,y_train,y_test,train=True)

met_score(ada,x_train,x_test,y_train,y_test,train=False)
```

```python
from sklearn.ensemble import RandomForestRegressor
ran_forst=RandomForestRegressor()
ran_forst.fit(x_train,y_train)
```

```python
met_score(ran_forst,x_train,x_test,y_train,y_test,train=True)

met_score(ran_forst,x_train,x_test,y_train,y_test,train=False)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gdbt_reg=GradientBoostingRegressor()
gdbt_reg.fit(x_train,y_train)
```

```python
met_score(gdbt_reg,x_train,x_test,y_train,y_test,train=True)

met_score(gdbt_reg,x_train,x_test,y_train,y_test,train=False)
```

```python
# XGB Model Initialization and build the model
from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
met_score(xgb,x_train,x_test,y_train,y_test,train=True)

met_score(xgb,x_train,x_test,y_train,y_test,train=False)
```
