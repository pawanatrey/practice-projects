```python
# Use sampling technique to manage imbalance data

from imblearn.over_sampling import SMOTE

sm=SMOTE(random_state=40,k_neighbors=5,sampling_strategy=0.80)
x_samp,y_samp=sm.fit_resample(x_scaler,y)
```

```python
# split the data

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.25,random_state=40)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# define function 

def met_score(mul_clf,x_train1,x_test1,y_train1,y_test1,train=True):
    if train:
        y_pred=mul_clf.predict(x_train1)
        print("***************Train Result************")
        print("Accuracy Score", accuracy_score(y_train1,y_pred))
    elif train==False:
        pred=mul_clf.predict(x_test1)
        print(pred)
        print("***************Test Result************")
        
        print("Confusion Matrix \n ", confusion_matrix(y_test1,pred))
        print("Accuracy Score", accuracy_score(y_test1,pred))
        
        print("Classification Report \n", classification_report(y_test1,pred))  
```

```python
# initialised KNN model

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
# call the function

met_score(knn,x_train,x_test,y_train,y_test,train=True)

met_score(knn,x_train,x_test,y_train,y_test,train=False)
```

```python
# initialised Decision model

from sklearn.tree import DecisionTreeClassifier
dec_tree=DecisionTreeClassifier()
dec_tree.fit(x_train,y_train)
```

```python
met_score(dec_tree,x_train,x_test,y_train,y_test,train=True)

met_score(dec_tree,x_train,x_test,y_train,y_test,train=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
ran_forst=RandomForestClassifier()
ran_forst.fit(x_train,y_train)
```

```python
met_score(ran_forst,x_train,x_test,y_train,y_test,train=True)

met_score(ran_forst,x_train,x_test,y_train,y_test,train=False)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada=AdaBoostClassifier()
ada.fit(x_train,y_train)
```

```python
met_score(ada,x_train,x_test,y_train,y_test,train=True)

met_score(ada,x_train,x_test,y_train,y_test,train=False)
```
