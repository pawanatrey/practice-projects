```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/FlipRoboTechnologies/ML-Datasets/main/Glass%20Identification/Glass%20Identification.csv")
df.head()
```

```python
# Delete id as it is no use

df=df.drop(['1'],axis=1)
df
```

```python
# Rename column

df.rename(columns = {'1.52101':'RI'}, inplace = True)
df.rename(columns = {'13.64':'Na'}, inplace = True)
df.rename(columns = {'4.49':'Mg'}, inplace = True)
df.rename(columns = {'1.10':'Al'}, inplace = True)
df.rename(columns = {'71.78':'Si'}, inplace = True)
df.rename(columns = {'0.06':'K'}, inplace = True)
df.rename(columns = {'8.75':'Ca'}, inplace = True)
df.rename(columns = {'0.00':'Ba'}, inplace = True)
df.rename(columns = {'0.00.1':'Fe'}, inplace = True)
df.rename(columns = {'1.1':'Glass'}, inplace = True)
```

```python
from sklearn.preprocessing import StandardScaler

x_scaler=StandardScaler()
x_scaler=x_scaler.fit_transform(x)
x_scaler
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.20,random_state=20)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,roc_auc_score,classification_report
```

```python
# define function 

def met_score(mul_clf,x_train1,x_test1,y_train1,y_test1,train=True):
    if train:
        y_pred=mul_clf.predict(x_train1)
        print("***************Train Result************")
        print("Accuracy Score", accuracy_score(y_train1,y_pred))
    elif train==False:
        pred=mul_clf.predict(x_test1)
        print(pred)
        print("***************Test Result************")
        
        print("Confusion Matrix \n ", confusion_matrix(y_test1,pred))
        print("Accuracy Score", accuracy_score(y_test1,pred))
        
        print("Classification Report \n", classification_report(y_test1,pred))  
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,roc_auc_score
```

```python
# Initialize Logistic Regression

lm=LogisticRegression()
lm.fit(x_train,y_train)
```

```python
# call the function

met_score(lm,x_train,x_test,y_train,y_test,train=True)
met_score(lm,x_train,x_test,y_train,y_test,train=False)
```

```python
# initialised KNN model

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
# call the function

met_score(knn,x_train,x_test,y_train,y_test,train=True)
met_score(knn,x_train,x_test,y_train,y_test,train=False)
```

```python
# initialised Decision model

from sklearn.tree import DecisionTreeClassifier
dec_tree=DecisionTreeClassifier()
dec_tree.fit(x_train,y_train)
```

```python
# call the function

met_score(dec_tree,x_train,x_test,y_train,y_test,train=True)
met_score(dec_tree,x_train,x_test,y_train,y_test,train=False)
```

```python
# call the function

met_score(ran_fort,x_train,x_test,y_train,y_test,train=True)
met_score(ran_fort,x_train,x_test,y_train,y_test,train=False)
```

```python
# ROC AUC Curve
from sklearn.metrics import roc_curve,auc

model ={'Logical Regression':lm,'Knn':knn,'Decision Tree':dec_tree,'Random Forest':ran_fort,'AdaBoost':ada,
                       'GBDT':gdbt_clas,'XGBDT':xgb,'Support Vector Machines':svc}
```

```python
plt.figure(figsize=(5,6))
for model_name,model_object in model.items():
    model_object.fit(x_train,y_train)
    y_pred=model_object.predict(x_test)
    fpr,tpr,thrashold=roc_curve(y_test,y_pred)
    roc_auc=auc(fpr,tpr)
    plt.plot(fpr,tpr,label='{} (AUC= {:.2f}'.format(model_name,roc_auc))
    
plt.plot([0,1],[0,1],linestyle='--',color='red')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristics (ROC) Curve')
plt.legend(loc='lower right')
    
#show the plot
plt.show()       
```
