```python
X=df.drop(['Survived'],axis=1)
Y=df.Survived
```

```python
from sklearn.preprocessing import StandardScaler

x_scaler=StandardScaler()
x_scaler=x_scaler.fit_transform(X)
x_scaler
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif=pd.DataFrame()
vif['Vif']=[variance_inflation_factor(x_scaler,i) for i in range(x_scaler.shape[1])]
vif['Feature']=X.columns
vif
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

x_train,x_test,y_train,y_test=train_test_split(x_scaler,Y,test_size=0.25,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_train)
print("***************Train Result************")
lr_train_score=round(accuracy_score(y_train,y_pred)*100,2)
print("Accuracy Score for Training Data",lr_train_score)
pred=lr.predict(x_test)
print("***************Test Result************")
lr_test_score=round(accuracy_score(y_test,pred)*100,2)
print("Accuracy Score for Test Data",lr_test_score)
print("Classification Report", classification_report(y_test,pred)) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_train)
print("***************Train Result************")
knn_train_score=round(accuracy_score(y_train,y_pred)*100,2)
print("Accuracy Score for Training Data",knn_train_score)
pred=knn.predict(x_test)
print("***************Test Result************")
knn_test_score=round(accuracy_score(y_test,pred)*100,2)
print("Accuracy Score for Test Data",knn_test_score)
print("Classification Report", classification_report(y_test,pred)) 
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_Tree=DecisionTreeClassifier()
dec_Tree.fit(x_train,y_train)
```

```python
y_pred=dec_Tree.predict(x_train)
print("***************Train Result************")
dec_train_score=round(accuracy_score(y_train,y_pred)*100,2)
print("Accuracy Score for Training Data",dec_train_score)
pred=dec_Tree.predict(x_test)
print("***************Test Result************")
dec_test_score=round(accuracy_score(y_test,pred)*100,2)
print("Accuracy Score for Test Data",dec_test_score)
print("Classification Report", classification_report(y_test,pred)) 
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada=AdaBoostClassifier()
ada.fit(x_train,y_train)
```

```python
y_pred=ada.predict(x_train)
print("***************Train Result************")
ada_train_score=round(accuracy_score(y_train,y_pred)*100,2)
print("Accuracy Score for Training Data",ada_train_score)
pred=ada.predict(x_test)
print("***************Test Result************")
ada_test_score=round(accuracy_score(y_test,pred)*100,2)
print("Accuracy Score for Test Data",ada_test_score)
print("Classification Report", classification_report(y_test,pred)) 
```

```python
y_pred=ran_forst.predict(x_train)
print("***************Train Result************")
rn_train_score=round(accuracy_score(y_train,y_pred)*100,2)
print("Accuracy Score for Training Data",rn_train_score)
pred=ran_forst.predict(x_test)
print("***************Test Result************")
rn_test_score=round(accuracy_score(y_test,pred)*100,2)
print("Accuracy Score for Test Data",rn_test_score)
print("Classification Report", classification_report(y_test,pred)) 
```

```python
# ROC AUC Curve
from sklearn.metrics import roc_curve,auc

model ={'Logical Regression':lr,'Knn':knn,'Decision Tree':dec_Tree,'AdaBoost':ada,'Random Forest':ran_forst,
                       'GBDT':gbdt,'XGBDT':xgb,'Support Vector Machines':svc}
```

```python
plt.figure(figsize=(5,6))
for model_name,model_object in model.items():
    model_object.fit(x_train,y_train)
    y_pred=model_object.predict(x_test)
    fpr,tpr,thrashold=roc_curve(y_test,y_pred)
    roc_auc=auc(fpr,tpr)
    plt.plot(fpr,tpr,label='{} (AUC= {:.2f}'.format(model_name,roc_auc))
    
plt.plot([0,1],[0,1],linestyle='--',color='grey')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristics (ROC) Curve')
plt.legend(loc='lower right')
    
#show the plot
plt.show()       
```
