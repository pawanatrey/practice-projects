```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/FlipRoboTechnologies/ML-Datasets/main/Grades/Grades.csv")
df.head()
```

```python
# delete seat no.
df=df.drop(['Seat No.'],axis=1)
```

```python
df.shape
```

```python
# Each column cantain the correct data type

df.info()
```

```python
# Number of missing values are there

df.isnull().sum()
```

```python
for i in df:
    print(df[i].value_counts())
```

```python
x=df.drop(['CGPA'],axis=1)
y=df.CGPA
```

```python
# Now we will check of mode value of each column

for i in x:
    print(x[i].mode())
```

```python
# Now impute the null values by most frequest(Mode)

from sklearn.impute import SimpleImputer
sim_imp=SimpleImputer(strategy='most_frequent')
fill_null=sim_imp.fit_transform(x)
x=pd.DataFrame(fill_null)
x
```

```python
# All Null values has imputed 

x.isnull().sum()
```

```python
x=pd.DataFrame(sim_imp.fit_transform(x),columns=['PH-121','HS-101','CY-105', 'HS-105/12', 'MT-111', 'CS-105', 'CS-106', 'EL-102', 
                                                   'EE-119','ME-107', 'CS-107','HS-205/20', 'MT-222', 'EE-222', 'MT-224', 'CS-210',
                                                   'CS-211', 'CS-203', 'CS-214', 'EE-217','CS-212', 'CS-215', 'MT-331','EF-303',
                                                   'HS-304', 'CS-301', 'CS-302', 'TC-383', 'MT-442','EL-332','CS-318', 'CS-306', 
                                                   'CS-312', 'CS-317', 'CS-403', 'CS-421', 'CS-406','CS-414','CS-419', 'CS-423', 
                                                   'CS-412'])
x
```

```python
from category_encoders import BinaryEncoder
bin_enc=BinaryEncoder()
x_tran=bin_enc.fit_transform(x)
x_tran
```

```python
x_tran.shape
```

```python
from sklearn.feature_selection import SelectKBest,f_classif
```

```python
best_features= SelectKBest(score_func=f_classif,k=30)
fit=best_features.fit(x_tran,y)  
df_scores=pd.DataFrame(fit.scores_)
df_scores
df_columns=pd.DataFrame(x_tran.columns)
df_columns
```

```python
# concatenate dataframes

features_score=pd.concat([df_columns,df_scores],axis=1)
features_score
```

```python
features_score.columns= ['Features_Name','Score']
X=features_score.nlargest(30,'Score') # It means find 50 largest features on the basis of score
print(X)
```

```python
# Selected Feature
new_x=x_tran[['CS-414_0','TC-383_0','EE-119_0','HS-304_0','CS-317_2','CS-406_0','EL-102_0','HS-205/20_3','CS-212_3','MT-224_2',
        'CS-106_1','PH-121_1','CS-211_2','CS-214_3','CS-203_0','CS-105_0','HS-101_1','EL-332_2','CS-317_0','CS-105_2',
        'MT-222_2','HS-105/12_3','MT-111_3','HS-205/20_0','CS-306_1','CS-214_2','CS-107_1','CS-312_2','MT-224_3','MT-442_0']]
new_x        
```

```python
from sklearn.preprocessing import StandardScaler

x_scaler=StandardScaler()
x_scaler=x_scaler.fit_transform(new_x)
x_scaler
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(new_x,y,test_size=0.20,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
print("*************Train Result************")
y_pred=regression.predict(x_train)
print("R2_Score", r2_score(y_train,y_pred))
print("Adjusted_R2_Score", regression.score(x_train,y_train))        
print("Mean Absolute Error", mean_absolute_error(y_train,y_pred))
print("Mean Square Error", mean_squared_error(y_train,y_pred))
print("Root Mean Square Error", np.sqrt(mean_squared_error(y_train,y_pred)))

pred=regression.predict(x_test)
print("***************Test Result************")
x1=r2_score(y_test,pred)
print("R2_Score", x1)
print("Adjusted_R2_Score", regression.score(x_test,y_test))  
print("Mean Absolute Error", mean_absolute_error(y_test,pred))
print("Mean Square Error", mean_squared_error(y_test,pred))
print("Root Mean Square Error", np.sqrt(mean_squared_error(y_test,pred)))    
```
